```python
ranks_win_rates = ['Herald Win Rate', 'Guardian Win Rate', 'Crusader Win Rate', 'Archon Win Rate', 'Legend Win Rate', 'Ancient Win Rate', 'Divine Win Rate', 'Immortal Win Rate']
top_10_per_rank = {}

for rank in ranks_win_rates:
    top_10_per_rank[rank] = df.nlargest(10, rank)
    
top_10_per_rank['Herald Win Rate']
```

```python
# Heatmap of win rates of heroes across different ranks

heatmap_data = df.pivot_table(index='Name', values=ranks_win_rates)
heatmap_data = heatmap_data.reindex(columns=ranks_win_rates)
heatmap_data = heatmap_data.sort_values(by='Immortal Win Rate', ascending=False)

fig, ax = plt.subplots(figsize=(15, 15))
fig.suptitle('Win Rates of Heroes Across Different Ranks', fontsize=20)

sns.heatmap(heatmap_data, cmap='coolwarm', ax=ax)
ax.set_xlabel('Rank', fontsize=15)
plt.xticks(rotation=0)
ax.set_ylabel('Hero', fontsize=15)

cbar = ax.collections[0].colorbar
cbar.set_label('Win Rate', fontsize=15)
cbar.ax.yaxis.set_label_position('right')


plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

```python
most_picked_heroes = {}

for hero in df['Name'].unique():
    total_picks = (
        df[df['Name'] == hero]['Herald Picks'].sum() +
        df[df['Name'] == hero]['Guardian Picks'].sum() +
        df[df['Name'] == hero]['Crusader Picks'].sum() +
        df[df['Name'] == hero]['Archon Picks'].sum() +
        df[df['Name'] == hero]['Legend Picks'].sum() +
        df[df['Name'] == hero]['Ancient Picks'].sum() +
        df[df['Name'] == hero]['Divine Picks'].sum() +
        df[df['Name'] == hero]['Immortal Picks'].sum()
    )
    most_picked_heroes[hero] = total_picks

most_picked_heroes = dict(sorted(most_picked_heroes.items(), key=lambda x: x[1], reverse=True))

most_picked_heroes = dict(list(most_picked_heroes.items())[:10])
```

```python
# Plot Top 10 most picked heroes

fig, ax = plt.subplots(figsize=(15, 8))
fig.suptitle('Top 10 Most Picked Heroes', fontsize=20)

sns.barplot(x=list(most_picked_heroes.keys()), y=list(most_picked_heroes.values()), ax=ax)
ax.set_xlabel('Hero', fontsize=15)
ax.set_ylabel('Total Picks', fontsize=15)
ax.tick_params(axis='x', rotation=45)

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

```python
# Pie Chart of the 10 most picked heroes

fig, ax = plt.subplots(figsize=(6, 6))
fig.suptitle('Top 10 Most Picked Heroes', fontsize=20)

ax.pie(list(most_picked_heroes.values()), labels=list(most_picked_heroes.keys()), autopct='%1.1f%%', startangle=90)
plt.show()
```
