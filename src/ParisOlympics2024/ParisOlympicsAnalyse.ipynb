```python
# Importing the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
from pathlib import Path
```

```python
# Load the data

csvs_path = 'paris olympics 2024 dataset updated/'
class DataFrames:
    def __init__(self):
        self.athletes = pd.read_csv(Path(csvs_path + 'athletes.csv'))
        self.coaches = pd.read_csv(Path(csvs_path + 'coaches.csv'))
        self.events = pd.read_csv(Path(csvs_path + 'events.csv'))
        self.medals = pd.read_csv(Path(csvs_path + 'medals.csv'))
        self.medals_total = pd.read_csv(Path(csvs_path + 'medals_total.csv'))
        self.schedules = pd.read_csv(Path(csvs_path + 'schedules.csv'))
        self.schedules_preliminary = pd.read_csv(Path(csvs_path + 'schedules_preliminary.csv'))
        self.teams = pd.read_csv(Path(csvs_path + 'teams.csv'))
        self.torch_route = pd.read_csv(Path(csvs_path + 'torch_route.csv'))
        self.venues = pd.read_csv(Path(csvs_path + 'venues.csv'))

dfs = DataFrames()

pd.DataFrame(dfs.__dict__.keys(), columns=['DataFrames'])
```

```python
country_distribution = dfs.athletes['country']
country_distribution = country_distribution.value_counts().reset_index()
country_distribution.columns = ['Country', 'Athletes']

# Plotting the distribution of athletes by country
plt.figure(figsize=(20, 10))
sns.barplot(x='Country', y='Athletes', data=country_distribution.head(25))
plt.title('Distribution of athletes by country')
plt.show()
```

```python
# Distribution of athletes by gender and disciplines

gender_category_distribution = dfs.athletes[['gender', 'disciplines']].groupby('gender').value_counts().reset_index(name='Athletes')

plt.figure(figsize=(20, 20))
sns.barplot(y='disciplines', x='Athletes', hue='gender', data=gender_category_distribution, errorbar=None)

plt.title('Distribution of athletes gender by disciplines')
plt.ylabel('Disciplines')
plt.legend(title='Gender')
plt.show()
```

```python
age_country_distribution = dfs.athletes[['birth_date', 'country']].copy()
age_country_distribution['birth_date'] = pd.to_datetime(age_country_distribution['birth_date'])
age_country_distribution['age'] = 2024 - age_country_distribution['birth_date'].dt.year
age_country_distribution = age_country_distribution[['country', 'age']]

age_country_average = age_country_distribution.groupby('country').mean().reset_index() 
age_country_average.columns = ['Country', 'Average Age']

plt.figure(figsize=(20, 10))
sns.kdeplot(data=age_country_average, fill=True, x='Average Age', multiple='stack')
plt.title('Athletes average age distribution')
plt.show()
```

```python
age_counts = age_country_distribution['age'].value_counts().reset_index()
age_counts.columns = ['Age', 'Athletes']
age_counts = age_counts.sort_values(by='Age')

plt.figure(figsize=(20, 10))
sns.barplot(x='Age', y='Athletes', data=age_counts, palette='viridis', hue='Age')
plt.title('Distribution of athletes by age')
plt.show()
```
